In the next step we perform the pre-processing steps for refining the dataset

```python
# Removing punctuations
data=train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)

# Renaming column names for ease of access
list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
data.columns= new_Index
data.head()
```

-----------------
In which we conter the Headlines to the lower case 

```python
# Convertng headlines to lower case
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(5)
```

```python
' '.join(str(x) for x in data.iloc[1,0:25])
```

```python
headlines = []
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[0]
```

-------------------------------------------------
In the next step we implement the Countvectorizer by RandomForestClassifier.

```python
### Countvectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
## implement BAG OF WORDS
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headlines)
```

```python
traindataset[0]
```

-------------------------------------------------------------------------
In which we implement the RandomForest Classifier

```python
# implement RandomForest Classifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

----------------------------------------------------------------
In the next cell we take prediction for the Test Dataset

```python
## Predict for the Test Dataset
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = countvector.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

---------------------------------------------------
In the next cell we show all the prediction in a array of list by implement the Countvectorizer and RandomForestClassifier.

```python
predictions
```

```python
plt.hist(predictions, label="Increase and Decrease in Stock after prediction",color="Purple")
plt.legend()
plt.ylabel("Count",color="purple")
plt.xlabel("Stock increase decrease",color="purple")
plt.title("In this graph we show the increasing and decreasing of the stock in the prediction",color="purple")
```

```python
test.loc[3723,:]
```

--------------------------------------------------------
In which we import the classification report,confusion matrix and the show the accuracy score

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

------------------------------------------------------------------------------------------------
In the next cell we show the confusion matrix,score of the classifier and the classification report

```python
matrix=confusion_matrix(test['Label'],predictions)
print("Matrix")
print(matrix)
score=accuracy_score(test['Label'],predictions)
print("RandomForest Classifier by countvectorizer Score")
print(score)
report=classification_report(test['Label'],predictions)
print("Report")
print(report)
```
